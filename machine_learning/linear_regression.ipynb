```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data=pd.read_csv('data/regression1.csv')
X=data.iloc[:,0].values
y=data.iloc[:,-1].values

mu=X.mean()
sigma=X.std()
Z=(X-mu)/sigma

plt.plot(Z,y,'o')
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
X1=Z.reshape(-1,1)
lr.fit(X1,y)
y_pred=lr.predict(X1)
print('slope: %.3f'%lr.coef_[0])
print('Intercept:%.3f'%lr.intercept_)
```

```python
x=np.linspace(-3,3,100)

plt.plot(Z,y,'o')
plt.plot(x,lr.intercept_+lr.coef_[0]*x)
```

```python
from sklearn.metrics import mean_squared_error
print('MSE: %.3f'%mean_squared_error(y,y_pred))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data=pd.read_csv('data/regression1.csv')
X=data.iloc[:,0].values
y=data.iloc[:,-1].values

mu=X.mean()
sigma=X.std()
Z=(X-mu)/sigma

plt.plot(Z,y,'o')
```

```python
from sklearn.preprocessing import PolynomialFeatures

pr = LinearRegression()
quadratic=PolynomialFeatures(degree=2)
X_quad = quadratic.fit_transform(X1)

pr.fit(X_quad,y)
y_quad_pred=pr.predict(X_quad)
print('theta1: %.3f'%pr.coef_[1])
print('theta2: %.3f'%pr.coef_[2])
print('Intercept:%.3f'%pr.intercept_)
```

```python
x=np.linspace(-3,3,100)

plt.plot(Z,y,'o')

plt.plot(x,lr.intercept_+lr.coef_[0]*x)

plt.plot(x,pr.intercept_+pr.coef_[1]*x+pr.coef_[2]*x**2)

print('MSE: %.3f'%mean_squared_error(y,y_quad_pred))
```
