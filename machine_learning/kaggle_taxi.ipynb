```python
import pandas as pd
import numpy as np
train_path = "./data/final_data/train.csv"
test_path = "./data/final_data/test.csv"
```

```python
df_tmp = pd.read_csv(train_path,nrows = 10000000)
df_tmp.head()
```

```python
df_test =  pd.read_csv(test_path)
df_test.head()
```

```python
df_tmp['pickup_datetime']=pd.to_datetime(df_tmp['pickup_datetime'],format='%Y-%m-%d %H:%M:%S UTC')
df_tmp.head()
```

```python
df_test['pickup_datetime']=pd.to_datetime(df_test['pickup_datetime'],format='%Y-%m-%d %H:%M:%S UTC')
df_test.head()
```

```python
data = [df_tmp,df_test]
for i in data:
    i['Year'] = i['pickup_datetime'].dt.year
    i['Month'] = i['pickup_datetime'].dt.month
    i['Date'] = i['pickup_datetime'].dt.day
    i['Day of Week'] = i['pickup_datetime'].dt.dayofweek
    i['Hour'] = i['pickup_datetime'].dt.hour
```

```python
df_tmp.head()
```

```python
df_tmp.dtypes
```

```python
train = df_tmp.drop(['key','pickup_datetime'], axis = 1)
train_dropna =  train.dropna()
test = df_test.drop(['key','pickup_datetime'], axis = 1)
```

```python
X_train = train_dropna.iloc[:,train_dropna.columns!='fare_amount']
y_train = train_dropna['fare_amount'].values
x_test = test
```

```python
np.any(np.isnan(X_train))
```

```python
np.any(np.isnan(y_train))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(X_train, y_train)
rf_predict = rf.predict(x_test)

submission = pd.read_csv('./data/final_data/sample_submission.csv')
submission['fare_amount'] = rf_predict
submission.to_csv('./data/final_data/submission_1000w.csv', index=False)
submission.head(20)
```
